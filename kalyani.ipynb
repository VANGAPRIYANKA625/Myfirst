```python
import matplotlib.pyplot as plt
data = {
    'SDID': [123, 456, 789, 147, 963],
    'SDNAME': ['Priyanka', 'Nikita', 'Kalyani', 'Vani', 'Goutam'],
    'YEAR2020': [90, 82, 80, 83, 65],
    'YEAR2021': [85, 95, 97, 98, 75],
    'YEAR2022': [79, 89, 88, 79, 69],
    'YEAR2023': [92, 75, 90, 82, 92]
}
x = data['SDID']
years = ['YEAR2020', 'YEAR2021', 'YEAR2022', 'YEAR2023']
plt.figure(figsize=(10, 6))
for year in years:
    plt.plot(x, data[year], marker='o', label=year)

plt.xlabel('SDNAME')
plt.ylabel('Scores')
plt.title('Scores by SDID over the Years')
plt.legend()
plt.grid(True)
plt.show()
```
